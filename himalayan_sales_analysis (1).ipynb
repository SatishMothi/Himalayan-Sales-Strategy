## Step 1: Import Libraries and Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('himalayan_sales_data.csv')
df.head()
```

## Step 2: Summary Statistics

```python
df = pd.read_csv('himalayan_sales_data.csv')
```

## Step 3: Sales by Region

```python
region_sales = df.groupby('Region')['Total_Sale'].sum().sort_values(ascending=False)
region_sales.plot(kind='bar', title='Total Sales by Region', ylabel='Sales ($)', xlabel='Region')
plt.tight_layout()
plt.show()
```

## Step 4: Product Popularity

```python
product_counts = df['Product'].value_counts()
sns.barplot(x=product_counts.values, y=product_counts.index)
plt.title('Units Sold by Product')
plt.xlabel('Units Sold')
plt.ylabel('Product')
plt.show()
```

## Step 5: Sales by Channel and Gender

```python
channel_gender_sales = df.pivot_table(values='Total_Sale', index='Sales_Channel', columns='Gender', aggfunc='sum')
channel_gender_sales.plot(kind='bar', stacked=True, title='Sales by Channel and Gender')
plt.ylabel('Sales ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
